```python
# 블랙잭 https://www.acmicpc.net/problem/2798
# 브루트 포스(Brute force)
# 조합 가능한 모든 문자열을 하나씩 대입해 보는 방식으로 암호를 해독하는 방법.

import itertools

N, M = map(int, input().split())
nums = list(map(int, input().split()))
c = itertools.combinations(nums,3)
c = list(c)

sums = []
for i in c:
    sums.append(sum(i))

sums.sort(reverse=True)

for i in sums:
    if i <= M:
        print(i)
        break
```

```python
N, M = map(int, input().split())
nums = list(map(int, input().split()))
sums = []

result = 0

for i in range(N):
    for j in range(i+1,N):
        for k in range(j+1,N):
            x = nums[i] + nums[j] + nums[k]
            if x <= M:
                sums.append(x)

print(max(sums))
```

### 2231

```python
def generate(N):
    lst = list(map(int,list(str(N))))
    s = sum(lst) + N
    return s

N = int(input())

a = 0
for i in range(1,N):
    if generate(i) == N:
        a = i
        break
print(a)
```

### 7568

```python
N = int(input())

lst = [(55,185),(58,183),(88,186),(60,175),(46,155)]

"""
for _ in range(N):
    x, y = map(int,input().split())
    lst.append((x,y))
"""
w = lst[:]  # 리스트 = 리스트 는 복사가 아닌 참조! (w를 수정하면 lst가 바뀌고, h를 수정하면 w,lst 둘다 바뀜)
h = lst[:]  # 리스트를 복사하려면 슬라이싱[:]을 이용하거나 .copy()메소드를 사용해야함
    
for i in range(N):
    for j in range(0,N-i-1):
        if w[j][0] < w[j+1][0]:
            temp = w[j+1]
            w[j+1] = w[j]
            w[j] = temp

            
for i in range(N):
    for j in range(0,N-i-1):
        if h[j][1] < h[j+1][1]:
            temp = h[j+1]
            h[j+1] = h[j]
            h[j] = temp


dic = {}
rank = 1
re = 0

for i in range(N):
    print(h[i], w[i])
    if h[i] == w[i]:
        rank += re
        dic[h[i]] = rank
        re = 0
        rank+=1
    elif h[i] in dic.keys():
        re += 1
        dic[w[i]] = rank
    elif w[i] in dic.keys():
        re += 1
        dic[h[i]] = rank
    else:
        re+=1
        dic[w[i]] = rank
        dic[h[i]] = rank
    print("rank : ",rank)
    print("re : ",re)
    print(dic)

for x in lst:
    print(dic[x],end=" ")
```

### 1018

```python
# https://www.acmicpc.net/problem/1018 : 체스판 다시 칠하기

counts = []
lst = []

N, M = map(int,input().split())

for _ in range(N):  # 체스판 입력받기
    lst.append(input())
    
for i in range(N-7):  # 가로 이동
    for j in range(M-7):  # 세로 이동
        count1 = 0
        count2 = 0
        for a in range(i,i+8):  # 가로인덱스 하나씩 증가
            for b in range(j,j+8):  # 세로인덱스 하나씩 증가
                if (a+b)%2 == 0:  # 대각선의 연관성
                    if lst[a][b] != 'B' : count1+=1 # 시작이 검정인 경우
                    if lst[a][b] != 'W' : count2+=1 # 시작이 흰색인 경우
                else:
                    if lst[a][b] != 'W' : count1+=1
                    if lst[a][b] != 'B' : count2+=1

        counts.append(count1)
        counts.append(count2)
            
print(min(counts))  # 가장 작은 값 출력
```

```python
def counter(m):
    count1 = 0
    count2 = 0
    for a in range(8):
        for b in range(8):
            if (a+b)%2 == 0:
                if m[a][b] != 'B' : count1+=1
                if m[a][b] != 'W' : count2+=1
            else:
                if m[a][b] != 'W' : count1+=1
                if m[a][b] != 'B' : count2+=1
                        
    return count1,count2

lst = []
temp = []

counts = []

N, M = map(int,input().split())

for i in range(N):
    lst.append(list(input()))
    
for i in range(N-7):
    for j in range(M-7):
        temp = lst[i:(-1)*(N-i-7)][j:(-1)*(M-j-7)]
        c1, c2 = counter(temp)
        counts.append(c1)
        counts.append(c2)
            
print(min(counts))
```
