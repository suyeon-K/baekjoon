### 12779

```python
from math import sqrt,gcd


a , b = map(int,input().split(" "))

bottom = b-a
top = int(sqrt(b)) - int(sqrt(a)) # 제곱근 개수 구하기
            
if top == 0:
    print(0)
else:
    div = gcd(top,bottom)
    print("{}/{}".format(top//div,bottom//div))
```

### 17271

```python
N,M = map(int,input().split(" "))

num = 0

for i in range(N//M+1):
    B = i
    A = N-M*i
    x = 1
    
    if B!=0:
        n = A + B
        for j in range(B):
            x *= (n-j)
        for j in range(B):
            x = x//(j+1)
    num += x
    
print(num%1000000007)
        
```
