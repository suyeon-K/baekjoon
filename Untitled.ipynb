```python
def cardReverse(nums,a,b): 
    n = nums[a-1:b] 
    n = n[::-1]
    nums[a-1:b] = n
    return nums
nums = list(range(1,21))
for _ in range(10):
    a, b = map(int, input().split())
    nums = cardReverse(nums,a,b)
print(' '.join(map(str, nums)))
```

```python
""" solution-2 직접 구현 """
nums = list(range(0,21))
for _ in range(10):
    a,b = map(int, input().split())
    for i in range((b-a)//2+1):
        nums[a+i], nums[b-i] = nums[b-i], nums[a+i]
nums.pop(0)
print(' '.join(map(str, nums)))
```
