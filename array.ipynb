### 1546

```python
N = int(input())

lst = input().split()
    
for i in range(N):
    lst[i] = int(lst[i])
    
max = lst[0]

for i in lst:
    if i > max:
        max = i

for i in range(N):
    lst[i] = round((lst[i]/max)*100,2)

sum = 0

for i in lst:
    sum += i

avg = round(sum/N,2)

print(avg)

```

### 4344

```python
C = int(input())

for i in range(C):
    
    x = input().split()
    N = x[0]
    
    x.remove(N)
    
    sum = 0
    
    for j in x:
        j = int(j)
        sum += j
    
    avg = sum/int(N)
    
    num = 0
    
    for j in x:
        j = int(j)
        if j > avg:
            num += 1
    
    rate = (num/int(N))*100
    
    print("{:.3f}%".format(rate))
```
