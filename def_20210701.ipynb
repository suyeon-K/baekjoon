```python
def is_self(n):
    ans = True
    if n<100:
        for i in range(10):
            for j in range(10):
                if n == i*11+j:
                    ans = False
    elif n < 1000:
        for i in range(10):
            for j in range(10):
                for k in range(10):
                    if n == i*101 + j*11 + k:
                        ans = False
    else:
        for l in range(10):
            for i in range(10):
                for j in range(10):
                    for k in range(10):
                        if n == l*1001 + i*101 + j*11 + k:
                            ans = False
    
    return ans
```

```python
def d(n):
    temp = str(n)
    x = n
    
    for i in temp:
        x+=int(i)
        
    return x

lst = []
for i in range(10000):
    lst.append(d(i))

lst_set = set(lst)
lst = list(lst_set)

for i in range(10000):
    if i not in lst:
        print(i)
```

### 1065

```python
def solve(x):
    
    count = 0
    
    if x < 100:
        count = x
    elif x <111:
        count = 99
    else:
        count = 99
        for i in range(111,x+1):
            temp = str(i)
            lst = []
            for j in temp:
                lst.append(int(j))
                
            div = lst[1]-lst[0]
            
            a = 0
            
            for k in range(len(lst)-2):
                if div != lst[k+2]-lst[k+1]:
                    break
                a += 1
                    
            if a == len(lst)-2:
                count += 1
                
    return count

N = int(input())
print(solve(N))
```
