### 2292

```python
# 백준 2292 (벌집) : https://www.acmicpc.net/problem/2292

# 13 = 1 + 6 + 6
# 58 = 1 + 6 + 12 + 18 + 21

N = int(input())

N -= 1
count = 1

while(True):
    if N <= 0:
        break
    N -= 6*count # 이동한 번호를 빼줌
    count += 1 #한 layer를 넘어갈 때 마다 6*i씩 증가

print(count)
```

### 2775

```python
# 부녀회장이 될테야! : https://www.acmicpc.net/problem/2775

T = int(input())  # 테스트 케이스 개수

for i in range(T):  #반복!
    k = int(input())  # 층
    n = int(input())  # 호
    
    nums = list(range(1,n+1))  # 이전 층에 1~n호 거주 인원 수를 담을 리스트

    x = 1  # 해당 호에 사는 사람 수

    for j in range(k): # 1층부터 k층까지 반복
        x = nums[0] # 각 층의 1호에 거주하는 사람 인원 수를 담음
        for l in range(1,n): # 1호부터 n호까지 구하기
            x += nums[l]
            nums[l]=x  # 인원수를 갱신

    print(x)
```

### 2839

```python
N = int(input())

num = N//5
while(True):
    if (N - num*5)%3 == 0:
        num += (N - num*5)//3
        break
    num -= 1
    if num <0:
        break

print(num)
```

### 1978

```python
def is_prime(x):
    if x == 1:
        return False
    if x == 2:
        return True
    for i in range(2,int(x**(1/2))+1):
        if x % i == 0:
            return False
    return True

N = int(input())
nums = list(map(int,input().split()))
count = 0

for num in nums:
    if (is_prime(num)):
        count += 1
print(count)
```
