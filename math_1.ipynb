### 2292

```python
# 백준 2292 (벌집) : https://www.acmicpc.net/problem/2292

# 13 = 1 + 6 + 6
# 58 = 1 + 6 + 12 + 18 + 21

N = int(input())

N -= 1
count = 1

while(True):
    if N <= 0:
        break
    N -= 6*count # 이동한 번호를 빼줌
    count += 1 #한 layer를 넘어갈 때 마다 6*i씩 증가

print(count)
```

# 1193

```python
N = int(input())

x = 1

while(True):
    N -= x
    
    if (N <= 0):
        break
        
    x += 1

a = N + x - 1
    
if x%2==1:
    top = x - a
    bottom = 1 + a
else:
    top = 1 + a
    bottom = x - a

print("{}/{}".format(top,bottom))
```

# 2869

```python
A, B, V = map(int, input().split())

V -= A
days = V//(A-B)

if V%(A-B) == 0:
    print(days+1)
else:
    print(days+2)
```

# 10250

### 2775

```python
# 부녀회장이 될테야! : https://www.acmicpc.net/problem/2775

T = int(input())  # 테스트 케이스 개수

for i in range(T):  #반복!
    k = int(input())  # 층
    n = int(input())  # 호
    
    nums = list(range(1,n+1))  # 이전 층에 1~n호 거주 인원 수를 담을 리스트

    x = 1  # 해당 호에 사는 사람 수

    for j in range(k): # 1층부터 k층까지 반복
        x = nums[0] # 각 층의 1호에 거주하는 사람 인원 수를 담음
        for l in range(1,n): # 1호부터 n호까지 구하기
            x += nums[l]
            nums[l]=x  # 인원수를 갱신

    print(x)
```

### 2839

```python
N = int(input())

num = N//5
while(True):
    if (N - num*5)%3 == 0:
        num += (N - num*5)//3
        break
    num -= 1
    if num <0:
        break

print(num)
```

```python
# 2839 설탕 배달 https://www.acmicpc.net/problem/2839
n = int(input()) # 설탕 무게
div_n = 0 # 설탕 무게를 5로 나눈 몫
while n >= 0: # 설탕 무게가 0이상일 때
    if n % 5 == 0: # 설탕 무게가 5의 배수일 때
        div_n += n // 5 # 5로 나눈 몫 더하기
        print(div_n) # 몫 출력
        break
    n -= 3 # 설탕 무게가 5의 배수가 될 때까지 3을 빼고
    div_n += 1 # 몫에 1씩 더하기(3으로 나누는 형식)
else:
    print(-1) # 도저히 안 떨어지면 -1 출력
```
