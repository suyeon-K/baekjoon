### 1712

### 2292

```python
# 백준 2292 (벌집) : https://www.acmicpc.net/problem/2292

# 13 = 1 + 6 + 6
# 58 = 1 + 6 + 12 + 18 + 21

N = int(input())

N -= 1
count = 1

while(True):
    if N <= 0:
        break
    N -= 6*count # 이동한 번호를 빼줌
    count += 1 #한 layer를 넘어갈 때 마다 6*i씩 증가

print(count)
```

# 1193

```python
N = int(input())

x = 1

while(True):
    N -= x
    
    if (N <= 0):
        break
        
    x += 1

a = N + x - 1
    
if x%2==1:
    top = x - a
    bottom = 1 + a
else:
    top = 1 + a
    bottom = x - a

print("{}/{}".format(top,bottom))
```

# 2869

```python
A, B, V = map(int, input().split())

V -= A
days = V//(A-B)

if V%(A-B) == 0:
    print(days+1)
else:
    print(days+2)
```

# 10250

```python
T = int(input())

for i in range(T):
    H, W, N = map(int,input().split())
    
    ho = N//H + 1
    floor = N%H
        
    print(floor*100 + ho)
```
