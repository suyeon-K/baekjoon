### 4948

```python
def is_prime(x):
    if x == 1:
        return False
    if x == 2:
        return True
    for i in range(2,int(x**(1/2))+1):
        if x % i == 0:
            return False
    return True

while(True):
    n = int(input())
    
    if n == 0:
        break
    
    cnt = 0
    if n == 1:
        cnt = 1
    elif n == 2:
        cnt = 2
    elif n%2 == 0:
        for i in range(n+1,2*n,2):
            if is_prime(i):
                cnt+=1
    else:
        for i in range(n,2*n,2):
            if is_prime(i):
                cnt+=1
    print(cnt)
```

### 9020

```python
def is_prime(x):
    if x == 1:
        return False
    if x == 2:
        return True
    for i in range(2,int(x**(1/2))+1):
        if x % i == 0:
            return False
    return True

T = int(input())

for _ in range(T):
    x = int(input())
    
    for i in range(x//2,0,-1):
        if is_prime(i):
            if is_prime(x-i):
                print(i, x-i)
                break
```

### 3009

```python
# 직사각형 나머지 모서리 좌표 구하기( https://www.acmicpc.net/problem/3009 )
# hint : 각 리스트에 2개씩 있어야함
# ex) (5,5)(5,7)(7,5)(7,7) -> x = [5,5,7,7] / y = [5,7,5,7]

xs = []
ys = []

for _ in range(3):
    x,y = map(int,input().split())
    if x not in xs:
        xs.append(x)
    else:
        xs.remove(x)
        
    if y not in ys:
        ys.append(y)
    else:
        ys.remove(y)

print(xs[0], ys[0])

```
