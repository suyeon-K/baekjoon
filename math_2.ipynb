### 4948

```python
# 4948: (https://www.acmicpc.net/problem/4948)

def is_prime(x):
    if x == 1: # 1은 소수도 합성수도 아님
        return False
    if x == 2: # 2는 소수인 유일한 짝수
        return True
    if x%2 ==0:
        return False
    for i in range(2,int(x**(1/2))+1): # 나눠지는 값이 없으면 소수!
        if x % i == 0:
            return False
    return True

while(True):
    n = int(input())
    
    if n == 0:
        break
    
    cnt = 0
    if n == 1:
        cnt = 1
    elif n == 2:
        cnt = 2
    elif n%2 == 0:
        for i in range(n+1,2*n,2):
            if is_prime(i):
                cnt+=1
    else:
        for i in range(n,2*n,2):
            if is_prime(i):
                cnt+=1
    print(cnt)
```

```python
# 오히려 입력 범위에 해당하는 모든 수들의 소수 여부를 먼저 구해줌으로써 시간을 단축시킬 수 있다.
# 에라토스테네스의 체(소수를 구하는 방법 중 하나)

N = 123456 * 2 + 1
sieve = [True] * N
for i in range(2, int(N**0.5)+1):
    if sieve[i]:
        for j in range(2*i, N, i):
            sieve[j] = False

def prime_cnt(val):
    cnt = 0
    for i in range(val + 1, val * 2 + 1):
        if sieve[i]:
            cnt += 1
    print(cnt)

while True:
    val = int(input())
    if val == 0:
        break
    prime_cnt(val)
```

### 9020

```python
# 골드바흐의 추측 : https://www.acmicpc.net/problem/9020

def is_prime(x):
    if x == 1: # 1은 소수도 합성수도 아님
        return False
    if x == 2: # 2는 소수인 유일한 짝수
        return True
    if x%2 ==0:
        return False
    for i in range(2,int(x**(1/2))+1): # 나눠지는 값이 없으면 소수!
        if x % i == 0:
            return False
    return True

T = int(input())

for _ in range(T):
    x = int(input())
    
    for i in range(x//2,0,-1): # 중간 수부터 소수인지 판별!
        if is_prime(i):
            if is_prime(x-i): # 차도 소수인지 판별!
                print(i, x-i)
                break
```

### 3009

```python
# 직사각형 나머지 모서리 좌표 구하기( https://www.acmicpc.net/problem/3009 )
# hint : 각 리스트에 2개씩 있어야함
# ex) (5,5)(5,7)(7,5)(7,7) -> x = [5,5,7,7] / y = [5,7,5,7]

xs = []
ys = []

for _ in range(3):
    x,y = map(int,input().split())
    if x not in xs:
        xs.append(x)
    else:
        xs.remove(x)
        
    if y not in ys:
        ys.append(y)
    else:
        ys.remove(y)

print(xs[0], ys[0])

```

### 1002

```python
T  = int(input())

for i in range(T):
    x_1, y_1, r_1, x_2, y_2, r_2 = input().split()

    d = (int(x_1) - int(x_2))**2 + (int(y_1) - int(y_2))**2  

    r_plus = (int(r_1) + int(r_2))**2
    r_minus = (int(r_1) - int(r_2))**2

    if d > r_plus :
        print(0)
    elif d == r_plus :
        print(1)
    elif d == r_minus :
        if x_1 == x_2 and y_1 == y_2 and r_1 == r_2:
            print(-1)
        else :
            print(1)
    elif d < r_minus :
        print(0)
    else:
        print(2)
```

```python
def 삼육구(num):
    num = list(str(num).split())
    for i, n in enumerate(num):
        if int(n)%3 == 0:
            num[i] = 'X'

    for n in num:
        print(n,end="")

n = int(input())

for i in range(1,n+1):
    print(삼육구(i))
```

```python
def triangle(list_n):
    if sum(list_n) == 0: 
        return False 
    else:
        list_n.sort(reverse=False)
        if (list_n[0] ** 2) + (list_n[1] ** 2) == list_n[2] ** 2: 
            return "right"
        else: 
            return "wrong"
while True:
    list_n = list(map(int, input().split()))
    if triangle(list_n) == False:
        break
    print(triangle(list_n))
```
