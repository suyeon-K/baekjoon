### 10872

```python
def factorial(N):
    if N == 0 or N == 1:
        return 1
    else:
        return N*factorial(N-1)

n = int(input())

print(factorial(n))
```

### 10870

```python
def f(n):
    if n ==0:
        return 0
    if (n == 2) or (n==1):
        return 1
    return f(n-1) + f(n-2)


N = int(input())
print(f(N))                    
```

```python
memo = {1:1,2:1}  # 메모화

def f(n):
    if n in memo:
        return memo[n] # 조기리턴
    output = f(n-1) + f(n-2)
    memo[n] = output
    return output
    
N = int(input())
print(f(N))     
```
