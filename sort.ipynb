### 2751

```python
N = int(input())

lst = [0 for i in range(N)]

for i in range(N):
    lst[i] = int(input())
    
for a in range(N-1):
    for b in range(0,N-a-1):
        if lst[b] > lst[b+1]:
            temp = lst[b]
            lst[b] = lst[b+1]
            lst[b+1] = temp

for i in lst:
    print(i)
```
