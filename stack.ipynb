```python
class stack:
    def __init__(self,N):
        self.stack_list = [0 for i in range(N)]
        self.idx = -1
        
    def size(self):
        return self.idx + 1
    
    def top(self):
        if self.size() == 0:
            return -1
        else:
            return self.stack_list[self.idx]
    
    def push(self,x):
        self.idx += 1
        self.stack_list[self.idx] = x
        
    
    def pop(self):
        if self.size() == 0:
            return -1
        else:
            self.idx -= 1
            return self.stack_list[self.idx + 1]
        
    def empty(self):
        if self.size() == 0:
            return 1
        else:
            return 0

        
N = int(input())
S = stack(N)

for _ in range(N):
    cmd = input().split()
    
    if cmd[0] == "push":
        S.push(int(cmd[-1]))
    elif cmd[0] == "pop":
        print(S.pop())
    elif cmd[0] == "top":
        print(S.top())
    elif cmd[0] == "empty":
        print(S.empty())
    elif cmd[0] == "size":
        print(S.size())
```

```python
class stack:
    def __init__(self):
        self.stack_list = []
        
    def size(self):
        return len(self.stack_list)
    
    def top(self):
        if self.size() == 0:
            return -1
        else:
            return self.stack_list[-1]
        
    def push(self,x):
        self.stack_list.append(x)
    
    def pop(self):
        if self.size() == 0:
            return -1
        else:
            return self.stack_list.pop()
        
    def empty(self):
        if self.size() == 0:
            return 1
        else:
            return 0
        
N = int(input())
S = stack()

for _ in range(N):
    cmd = input().split()
    
    if cmd[0] == "push":
        S.push(int(cmd[-1]))
    elif cmd[0] == "pop":
        print(S.pop())
    elif cmd[0] == "top":
        print(S.top())
    elif cmd[0] == "empty":
        print(S.empty())
    elif cmd[0] == "size":
        print(S.size())
```
