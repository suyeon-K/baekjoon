```python
class stack:
    def __init__(self,N):
        self.stack_list = [0 for i in range(N)]
        self.idx = -1
        
    def size(self):
        return self.idx + 1
    
    def top(self):
        if self.size() == 0:
            return -1
        else:
            return self.stack_list[self.idx]
    
    def push(self,x):
        self.idx += 1
        self.stack_list[self.idx] = x
        
    
    def pop(self):
        if self.size() == 0:
            return -1
        else:
            self.idx -= 1
            return self.stack_list[self.idx + 1]
        
    def empty(self):
        if self.size() == 0:
            return 1
        else:
            return 0

        
N = int(input())
S = stack(N)

for _ in range(N):
    cmd = input().split()
    
    if cmd[0] == "push":
        S.push(int(cmd[-1]))
    elif cmd[0] == "pop":
        print(S.pop())
    elif cmd[0] == "top":
        print(S.top())
    elif cmd[0] == "empty":
        print(S.empty())
    elif cmd[0] == "size":
        print(S.size())
```

```python
class stack:
    def __init__(self):
        self.stack_list = []
        
    def size(self):
        return len(self.stack_list)
    
    def top(self):
        if self.size() == 0:
            return -1
        else:
            return self.stack_list[-1]
        
    def push(self,x):
        self.stack_list.append(x)
    
    def pop(self):
        if self.size() == 0:
            return -1
        else:
            return self.stack_list.pop()
        
    def empty(self):
        if self.size() == 0:
            return 1
        else:
            return 0
        
N = int(input())
S = stack()

for _ in range(N):
    cmd = input().split()
    
    if cmd[0] == "push":
        S.push(int(cmd[-1]))
    elif cmd[0] == "pop":
        print(S.pop())
    elif cmd[0] == "top":
        print(S.top())
    elif cmd[0] == "empty":
        print(S.empty())
    elif cmd[0] == "size":
        print(S.size())
```

### 플러스 문제

```python
"""
<유효한 괄호>  https://leetcode.com/problems/valid-parentheses/
괄호로 된 입력값이 올바른지 판별하라.
"""
class Solution(object):
    def isValid(self, s):
        stack = []
        table = {
            # key-value
            ')':'(',
            '}':'{',
            ']':'[',
        }
        for i in s:
            if i not in table:  # 키값이 아니라면 , 딕셔너리만 쓰면 키값이 리스트로 반환
                stack.append(i)
            elif not stack or table[i] != stack.pop():  # 키값이라면 , stack이 비어있으면 0이고 그러면 not 0 으로 1이 되서 False를 리턴
                return False
        return len(stack) == 0
s = "()[]"
a = Solution
print(a.isValid(a,s))
"""
<해설>
table 안에 주어진 문자 1개가 들어가 있지 않다면 stack에 추가한다.
반대로, table 안에 문자가 들어있을 때 stack이 비어 있거나 table의 value값과 stack의 마지막 값이 일치하는지 확인한다.
두 조건 중 하나라도 일치한다면, False를 반환한다. s 안의 모든 문자들을 돌고 나서 stack이 비어있다면 True를 반환한다.
"""
```

```python
class Solution(object):
    def isValid(self, s):
        stack = []
        table = {
            # key-value
            ')':'(',
            '}':'{',
            ']':'[',
        }
        for i in s:
            if i not in table.keys():  # 키값이 아니라면
                stack.append(i)
            elif len(stack) == 0 or table[i] != stack.pop():  # 키값이라면
                return False
        return len(stack) == 0
s = "()[]"
a = Solution
print(a.isValid(a,s))
```

### 9012

```python
# 9012  https://www.acmicpc.net/problem/9012

def is_VPS(st):
    x = st
    lst = []
    
    for i in x:
        if i in ["(","[","{"]:
            lst.append(i)
        else:
            if len(lst) == 0:
                return "NO"
            
            temp = lst.pop()
            
            if (i == ")"):
                if temp != "(":
                    return "NO"
            elif (i == "]"):
                if temp != "[":
                    return "NO"
            elif (i == "}"):
                if temp != "{":
                    return "NO"
    
    if len(lst) != 0:
        return "NO"
    
    return "YES"

N = int(input())
for _ in range(N):
    st = input()
    print(is_VPS(st))
```

### 4949

```python
# 4949 https://www.acmicpc.net/problem/4949

def is_VPS(st):
    x = st
    lst = []
    
    for i in x:
        if i not in ["(","[","]",")"]:
            continue
        if i in ["(","["]:
            lst.append(i)
        else:
            if len(lst) == 0:
                return "no"
            
            temp = lst.pop()
            
            if (i == ")"):
                if temp != "(":
                    return "no"
            elif (i == "]"):
                if temp != "[":
                    return "no"
    
    if len(lst) != 0:
        return "no"
    
    return "yes"

while(True):
    st = input()
    if st == ".":
        break
    print(is_VPS(st))
```

```python
# import sys
# input = sys.stdin.readline

N = int(input())

lst = list(map(int,input().split()))

ans = [-1]*N

for i in range(N-1):
    if lst[i] >= max(lst[i+1:]):
        continue
    for j in lst[i+1:]:
        if j>lst[i]:
            ans[i] = j
            break

for i in ans:
    print(i, end=" ")
```

```python
def cardReverse(nums,a,b): 
    n = nums[a-1:b] 
    n = n[::-1]
    nums[a-1:b] = n
    
    return nums

nums = list(range(1,21))
for _ in range(10):
    a, b = map(int, input().split())
    nums = cardReverse(nums,a,b)
print(' '.join(map(str, nums)))
```
