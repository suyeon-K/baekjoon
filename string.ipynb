### 1157

```python
s = input().upper()

s = list(s)

s_unique = list(set(s))
cnt = []

for i in s_unique:
    cnt.append(s.count(i))


max = cnt[0]
num = 0
idx = 0


for i,v in enumerate(cnt):
    if v > max:
        max = v
        num = 1
        idx = i
    elif v == max:
        num += 1

        
if num == 1:
    print(s_unique[idx])
elif num > 1:
    print("?")
```

### 5622

```python
s = input()
num = []

for i in s:
    if i in ["A","B","C"]:
        num.append(2)
    elif i in ["D","E","F"]:
        num.append(3)
    elif i in ["G","H","I"]:
        num.append(4)
    elif i in ["J","K","L"]:
        num.append(5)
    elif i in ["M","N","O"]:
        num.append(6)
    elif i in ["P","Q","R","S"]:
        num.append(7)
    elif i in ["T","U","V"]:
        num.append(8)
    elif i in ["W","X","Y","Z"]:
        num.append(9)

sec = 0
for n in num:
    sec += n+1

print(sec)
```

### 2941

```python
# 백준 2941
x = input()

# 문제풀이 의도 : 두글자 이상의 글자모음이 한 글자처럼 취급 되는 경우를 'a'라는 한글자로 바꿔줌
# 이때 z= 먼저 바꿔줄 경우 dz= 가 한글자가 아닌 da 이렇게 두글자 처리가 되므로 dz=먼저 a로 바꿔줌

x = x.replace('c=',"a")
x = x.replace('c-',"a")
x = x.replace('dz=',"a")
x = x.replace('d-=',"a")
x = x.replace('lj',"a")
x = x.replace('nj',"a")
x = x.replace('s=',"a")
x = x.replace('z=',"a")

# 한 글자 모음으로 이뤄진 문자열의 길이 출력
print(len(x))
```

### 1316

```python
N = int(input())
count = 0
for i in range(N):
    x = input()
    lst = []
    count += 1
    
    for idx, c in enumerate(list(x)):
        if c not in lst:
            lst.append(c)
        else:
            if c != x[idx-1]:
                count -= 1
                break

print(count)
```
